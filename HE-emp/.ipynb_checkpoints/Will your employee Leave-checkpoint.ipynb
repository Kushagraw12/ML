```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(tds.Attrition_rate, bins = 20)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
tds["Gender"] = le.fit_transform(tds["Gender"])
```

```python
tds["Relationship_Status"] = le.fit_transform(tds["Relationship_Status"])
tds["Unit"] = le.fit_transform(tds["Unit"])
tds["Decision_skill_possess"] = le.fit_transform(tds["Decision_skill_possess"])
tds["Compensation_and_Benefits"] = le.fit_transform(tds["Compensation_and_Benefits"])
tds["Hometown"] = le.fit_transform(tds["Hometown"])
```

```python
y = tds["Attrition_rate"]
y = y.astype(float)
```

```python
X = tds.drop(["Attrition_rate", "Employee_ID"], axis = 1)
X.columns.size
```

```python
X.head()
```

```python
from sklearn.linear_model import LassoCV
```

```python
reg = LassoCV()
reg.fit(X, y)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
corr = tds.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
coef = coef.abs()
imp_coef = coef.sort_values()
import matplotlib as mtp
mtp.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

```python
# Correlation Matrix
X.columns.size
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
## Model Building
```

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Dense(5, input_dim = 11, kernel_initializer = 'normal', activation = 'linear'))
model.add(Dropout(0.2))
model.add(Dense(5, kernel_initializer = 'normal', activation = 'linear'))
model.add(Dropout(0.2))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
opt = SGD(lr=0.01, momentum=0.9)
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 15, batch_size = 10)
```

```python
model.evaluate(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
from math import sqrt
```

```python
y_pred = model.predict(x_test)
```

```python
dl_rmse = sqrt(mse(y_test, y_pred))
print(dl_rmse)
```

REGRESSION MODEL

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(x_train, y_train)
```

```python
rf_pred = rf.predict(x_test)
```

```python
rf_rmse = sqrt(mse(y_test, rf_pred))
print(rf_rmse)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
lr_rmse = sqrt(mse(y_test, lr_pred))
print(lr_rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor  
```

```python
dt = DecisionTreeRegressor(random_state = 0)
dt.fit(x_train, y_train)
```

```python
dt_pred = dt.predict(x_test)
dt_rmse = sqrt(mse(y_test, dt_pred))
print(dt_rmse)
```

```python
plt.plot(y_test, label='Actual')
plt.plot(dt_pred, label='Predicted')
plt.title("Decision Tree")
plt.legend(frameon=True)
plt.show()
```

```python
from sklearn import datasets, ensemble
params = {'n_estimators': 100,
          'max_depth': 4,
          'learning_rate': 0.01,
          'loss': 'ls'}

gradient_boosting = ensemble.GradientBoostingRegressor(**params)
gradient_boosting.fit(x_train, y_train)

gb_predictions = gradient_boosting.predict(x_test)
gb_rmse = sqrt(mse(y_test, gb_predictions))

print("Gradient Boosting Regression:", 100 * max(0,1 - gb_rmse))
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
alphas = [0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 1]
elasticnet = ElasticNetCV(alphas = alphas).fit(x_train, y_train)
```

```python
score = elasticnet.score(x_train, y_train)
print(score)
```

```python
en_pred = elasticnet.predict(x_test)
```

```python
en_rmse = sqrt(mse(y_test, en_pred))
print(100 * max(0, 1 - en_rmse))
```

```python
lasso = LassoCV(alphas=alphas,normalize=True)
lasso.fit(x_train, y_train)
lasso_predictions = lasso.predict(x_test)
lasso_rmse = sqrt(mse(y_test, lasso_predictions))

print("Lasso Regression:", 100 * max(0,1 - lasso_rmse))
```

```python
from sklearn.linear_model import RidgeCV

ridge = RidgeCV(alphas=alphas,normalize=True)
ridge.fit(x_train, y_train)
rr_predictions = ridge.predict(x_test)
rr_rmse = sqrt(mse(y_test, rr_predictions))

print("Ridge Regression:", 100 * max(0,1 - rr_rmse))
```

```python
print("Score for different Regression models:")
print("Random Forest:", 100 * max(0,1 - rf_rmse))
print("Linear Regression:", 100 * max(0,1 - lr_rmse))
print("Decision Tree:", 100 * max(0,1 - dt_rmse))
print("Gradient Boosting Regression:", 100 * max(0,1 - gb_rmse))
print("Ridge Regression:", 100 * max(0,1 - rr_rmse))
print("Lasso Regression:", 100 * max(0,1 - lasso_rmse))
print("Elastic Net:", 100 * max(0,1 - en_rmse))
print("Deep Learning:", 100 * max(0,1 - dl_rmse))
```
